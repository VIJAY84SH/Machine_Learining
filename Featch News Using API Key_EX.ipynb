```python
main_url = "https://newsapi.org/v2/top-headlines?country=in&apiKey="+api_key
```

```python
def news(main_url):                    
    r = requests.get(main_url)        # request to the URL To get the Data
    data = json.loads(r.content)      # data will load into JSON format
    df = pd.DataFrame(columns=['id','source','articles','description'])  # created empty Dataframe to store articles
    for i in range(len(data['articles'])):
        df.loc[i,'id']= i
        df.loc[i,'source']= data['articles'][i]['source']['name']
        df.loc[i,'articles']= data['articles'][i]['title']
        df.loc[i,'description']= data['articles'][i]['description']
    return df
 
```
