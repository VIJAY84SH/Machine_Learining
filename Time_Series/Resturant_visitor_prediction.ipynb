```python
import pmdarima
from pmdarima import auto_arima
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
auto_arima(rv.total, seasonal=True, m = 7)
```

```python
rv.shape
```

```python
rv_train = rv.iloc[0:450] 
rv_test = rv.iloc[450:]
```

```python
model_sarima = SARIMAX(rv_train.total,order=(1,0,0), seasonal_order=(1,0,1,7))
```

```python
results_rv = model_sarima.fit()
```

```python
pred_rv = results_rv.predict(start = 450, end = 478)
#pred
```

```python
pred_rv
```

### Evaluate

```python
import numpy as np
```

```python
rv_test['Forecasted_sarima'] = np.round(pred_rv)
rv_test['error_sarima'] = pred_rv - rv_test.total
```

```python
rv_test.loc[:,['holiday','total','Forecasted_sarima']]
```

```python
mape_rv_sarima = np.mean(np.abs(rv_test['error_sarima']*100/rv_test.total))
```

```python
mape_rv_sarima
```

```python
mse_rv_sarima = np.mean(np.square(rv_test['error_sarima']))
```

```python
mse_rv_sarima
```

### Lets build SARIMAX model and compare with SARIMA

```python
model_sarima_rv = SARIMAX(rv_train.total, order=(1,0,0),seasonal_order=(1,0,1,7),
                         exog = rv_train[['holiday']])
```

```python
result_sarimax = model_sarima_rv.fit()
```

```python
pred_sarimax = result_sarimax.predict(start = 450, end = 477, exog= rv_test[['holiday']])
```

```python
pred_sarimax
```

```python
rv_test['predicted_SarimaX'] = pred_sarimax
```

```python
rv_test.loc[:,['holiday','total','Forecasted_sarima','error_sarima','predicted_SarimaX']]
```

### Holidays

```python
rv_test[rv_test.holiday==1].loc[:,['holiday','total','Forecasted_sarima','error_sarima','predicted_SarimaX']]
```
