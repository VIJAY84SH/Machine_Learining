```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('Insurance_linear_Regg.xlsx')
```

```python
df
```

### WE will predict Premium for Age 21 and 51

```python
plt.scatter(df.Age,df.Premium)
```

### Best Fit-line or Regression Line

```python
sns.lmplot(x = 'Age', y = 'Premium', data = df)
```

### Premium = M *Age +C

#### M - Slope....C- Intercepy

```python
from sklearn import linear_model
```

```python
reg = linear_model.LinearRegression()
```

```python
df.Age
```

```python
reg.fit(df[['Age']],df['Premium'])   # created a model
```

### Prediction for Age 21 and 51 as bellow

```python
reg.predict([[21]])   # for age 21 premium will be 13880
```

```python
reg.predict([[51]])    # for age 51 premium will be 67280
```

```python
reg.predict([[25]]) 
```

### y = mx+c

#### Premium = M *Age +C

```python
reg.coef_   # M
```

```python
reg.intercept_  # C
```
