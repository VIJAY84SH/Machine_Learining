### Remove Outlier

```python
from scipy.stats import zscore

z_scores = zscore(df)
no_outliers = df[(z_scores < 3).all(axis=1)]  # Keep only rows within 3 standard deviations
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
```

```python
scaled_df = scalar.fit_transform(df)
```

```python
scaled_df
```

```python
from sklearn.decomposition import PCA
pca = PCA()
pca_df = pca.fit_transform(scaled_df)

# Calculate cumulative explained variance
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)
cumulative_variance

# Find the number of components to retain 95% variance
num_components = np.argmax(cumulative_variance >= 0.95) + 1
num_components

# Apply PCA with chosen components
final_pca = PCA(n_components=num_components)
pca_df = final_pca.fit_transform(scaled_df)
scaled_df.shape, pca_df.shape
```

```python
from sklearn.cluster import KMeans
```

```python
from sklearn.cluster import KMeans

inertia_values = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(pca_df)
    inertia_values.append(kmeans.inertia_)

# Plot the elbow curve
plt.plot(range(1, 11), inertia_values, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Curve')
plt.show();
```

```python
from sklearn.metrics import silhouette_score
silhouette_scores = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(pca_df)
    silhouette_scores.append(silhouette_score(pca_df, labels))

# Plot silhouette scores
plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score Method')
plt.show()

# Build K-means model with the chosen k
optimal_k = np.argmax(silhouette_scores) + 2  # +2 because range starts from 2
kmeans_final = KMeans(n_clusters=optimal_k, random_state=42)
labels_final = kmeans_final.fit_predict(pca_df)

# Bar plot showing the number of observations in each cluster
plt.bar(range(optimal_k), np.bincount(labels_final))
plt.xlabel('Cluster')
plt.ylabel('Number of Observations')
plt.title('Number of Observations in Each Cluster')
plt.show()
```

### Conclusion

### K vlaue for Kmean Clustering using Elbow Method is 3

### K Value for Kmeans Clustering using silhouette score is 9
